# Statistics_Basics


1. What is the difference between descriptive statistics and inferential
statistics? Explain with examples.
   - Descriptive statistics summarize and organize data using measures like mean, median, mode, and standard
deviation. Example: calculating the average marks of students.
Inferential statistics use a sample to make conclusions about a population. Example: predicting the average
height of all students using a random sample.
2. What is sampling in statistics? Explain the differences between random
and stratified sampling.
   - Sampling means selecting a subset from a population for analysis.
Random sampling: Every individual has an equal chance of selection.
Stratified sampling: The population is divided into groups (strata) and samples are taken from each group
proportionally.
3. Define mean, median, and mode. Explain why these measures of central
tendency are important.
   - Mean: Average value.
Median: Middle value in ordered data.
Mode: Most frequent value.
These are important because they describe the central tendency of data.
4. Explain skewness and kurtosis. What does a positive skew imply about the
data?
   - Skewness measures asymmetry in data distribution. A positive skew means the tail is longer on the right
side.
Kurtosis measures the 'peakedness' of data. High kurtosis indicates heavier tails (more outliers)




```python
#Q5. Implement a Python program to compute the mean, median, and mode of a given list of numbers.
# numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


import statistics as st


numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
mean = st.mean(numbers)
median = st.median(numbers)
mode = st.mode(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)


```

```python
#Q6. Compute the covariance and correlation coefficient between the following  two datasets provided as lists in Python:
# list_x = [10, 20, 30, 40, 50]
# list_y = [15, 25, 35, 45, 60]

import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

covariance = np.cov(list_x, list_y)[0, 1]

correlation = np.corrcoef(list_x, list_y)[0, 1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
#Q7. Write a Python script to draw a boxplot for the following numeric list and  identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

import matplotlib.pyplot as plt
import numpy as np

plt.boxplot(data)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Outliers:", outliers)
```

```python
#Q8. You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and  daily sales.
# ‚óè Explain how you would use covariance and correlation to explore this relationship.
# ‚óè Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

import numpy as np


covariance = np.cov(advertising_spend, daily_sales)[0, 1]
correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

8. Explain how you would use covariance and correlation to explore this relationship.

   - Explanation:

To understand the relationship between advertising spend and daily sales, we use:

üîπ Covariance:

Covariance shows how two variables change together.

If both increase together ‚Üí positive covariance

If one increases while the other decreases ‚Üí negative covariance

But covariance does not show strength, only direction.

üîπ Correlation:

Correlation standardizes covariance to a value between ‚Äì1 and +1.

+1 ‚Üí Perfect positive relationship

0 ‚Üí No relationship

‚Äì1 ‚Üí Perfect negative relationship

It shows how strongly advertising spend is related to sales.

```python
#Q9. Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.
# ‚óè Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you‚Äôd use.
# ‚óè Write Python code to create a histogram using Matplotlib for the survey data:

import matplotlib.pyplot as plt
import statistics as st

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

mean = st.mean(survey_scores)
median = st.median(survey_scores)
mode = st.mode(survey_scores)
stdev = st.stdev(survey_scores)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Standard Deviation:", round(stdev, 2))

plt.hist(survey_scores, bins=6, color='skyblue', edgecolor='black')
plt.title('Customer Satisfaction Distribution')
plt.xlabel('Survey Scores (1‚Äì10)')
plt.ylabel('Frequency')
plt.show()
```
